# US Flag Vessel Analysis - Exploratory Analysis

This notebook provides interactive exploration of vessel data and MSC analysis results.

```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Add project root to path
sys.path.insert(0, '..')

# Set visualization style
sns.set_style('whitegrid')
sns.set_palette('Set2')
plt.rcParams['figure.figsize'] = (12, 6)

print("Setup complete")
```

## Load Processed Data

```python
# Load vessel inventory
vessel_file = Path('../data/processed/vessel_inventory.csv')
if vessel_file.exists():
    vessel_df = pd.read_csv(vessel_file)
    print(f"Loaded {len(vessel_df)} vessels")
    print(f"Columns: {list(vessel_df.columns)}")
else:
    print("No processed vessel data found. Run main.py --process first.")
    vessel_df = pd.DataFrame()
```

```python
# Load operator directory
operator_file = Path('../data/processed/operator_directory.csv')
if operator_file.exists():
    operator_df = pd.read_csv(operator_file)
    print(f"Loaded {len(operator_df)} operators")
else:
    print("No operator data found")
    operator_df = pd.DataFrame()
```

## Vessel Data Overview

```python
# Display first few records
if not vessel_df.empty:
    display(vessel_df.head())
```

```python
# Summary statistics
if not vessel_df.empty:
    print("\nSummary Statistics:")
    print(vessel_df.describe())
```

## Fleet Composition Analysis

```python
# Vessels by type
if not vessel_df.empty and 'vessel_type' in vessel_df.columns:
    type_counts = vessel_df['vessel_type'].value_counts().head(10)
    
    plt.figure(figsize=(12, 6))
    type_counts.plot(kind='barh')
    plt.title('Top 10 Vessel Types', fontsize=14, fontweight='bold')
    plt.xlabel('Count')
    plt.ylabel('Vessel Type')
    plt.tight_layout()
    plt.show()
```

```python
# Age distribution
if not vessel_df.empty and 'build_year' in vessel_df.columns:
    current_year = pd.Timestamp.now().year
    vessel_df['age'] = current_year - vessel_df['build_year']
    
    plt.figure(figsize=(12, 6))
    vessel_df['age'].dropna().hist(bins=30, edgecolor='black')
    plt.title('Vessel Age Distribution', fontsize=14, fontweight='bold')
    plt.xlabel('Age (years)')
    plt.ylabel('Count')
    plt.axvline(vessel_df['age'].median(), color='red', linestyle='--', 
                label=f"Median: {vessel_df['age'].median():.1f} years")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

## MSC Vessel Analysis

```python
# Load MSC vessels
msc_file = Path('../data/processed/msc_vessels.csv')
if msc_file.exists():
    msc_df = pd.read_csv(msc_file)
    print(f"Loaded {len(msc_df)} MSC vessels")
    display(msc_df.head())
else:
    print("No MSC vessel data found. Run main.py --analyze first.")
    msc_df = pd.DataFrame()
```

```python
# MSC vessels by category
if not msc_df.empty and 'msc_category' in msc_df.columns:
    category_counts = msc_df['msc_category'].value_counts()
    
    plt.figure(figsize=(10, 8))
    plt.pie(category_counts.values, labels=category_counts.index, 
            autopct='%1.1f%%', startangle=90)
    plt.title('MSC Vessels by Category', fontsize=14, fontweight='bold')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

## Operator Analysis

```python
# Top operators
if not operator_df.empty:
    print("Top 10 Operators by Vessel Count:")
    display(operator_df.nlargest(10, 'vessel_count')[['operator_name', 'vessel_count', 'vessel_types']])
```

```python
# Civilian vs military operated
if not operator_df.empty and 'is_civilian' in operator_df.columns:
    civilian_stats = operator_df.groupby('is_civilian')['vessel_count'].sum()
    
    plt.figure(figsize=(8, 8))
    labels = ['Military Operated', 'Civilian Operated']
    plt.pie(civilian_stats.values, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title('Vessels by Operator Type', fontsize=14, fontweight='bold')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

## Norfolk Vessels

```python
# Load Norfolk vessels
norfolk_file = Path('../data/processed/norfolk_vessels.csv')
if norfolk_file.exists():
    norfolk_df = pd.read_csv(norfolk_file)
    print(f"Loaded {len(norfolk_df)} Norfolk-area vessels")
    display(norfolk_df.head())
else:
    print("No Norfolk vessel data found")
    norfolk_df = pd.DataFrame()
```

```python
# Norfolk vessels by status
if not norfolk_df.empty and 'status_standardized' in norfolk_df.columns:
    status_counts = norfolk_df['status_standardized'].value_counts()
    
    plt.figure(figsize=(10, 6))
    status_counts.plot(kind='bar')
    plt.title('Norfolk MSC Vessels by Status', fontsize=14, fontweight='bold')
    plt.xlabel('Status')
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
